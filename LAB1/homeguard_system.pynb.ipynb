```python
def process_reading(sensor, reading_value):
    alerts = []
    
    if sensor["type"] == "temperature" and reading_value > sensor["threshold"]:
        alerts.append(create_alert("MEDIUM", "Temperature High", sensor["sensor_id"], datetime.now().strftime("%H:%M:%S")))
    elif sensor["type"] == "door" and reading_value == "OPEN":
        alerts.append(create_alert("HIGH", "Door opened!", sensor["sensor_id"], datetime.now().strftime("%H:%M:%S")))
    elif sensor["type"] == "motion" and reading_value == True:
        alerts.append(create_alert("CRITICAL", "You are being robbed!", sensor["sensor_id"], datetime.now().strftime("%H:%M:%S")))
    
    return alerts
```

```python
# A class is a blueprint - like a cookie cutter that makes sensor "cookies"
class Sensor:
    
    # This runs automatically when you create a new sensor
    def __init__(self, sensor_id, location, sensor_type, threshold=None):
        # "self" means "this specific sensor" - storing its properties
        self.id = sensor_id
        self.location = location
        self.type = sensor_type
        self.threshold = threshold
        self.current_value = None  # No reading yet
    
    # Generates a random reading based on what type of sensor this is
    def read(self):
        if self.type == "temperature":
            # Random temperature between 30-100, rounded to 1 decimal
            self.current_value = round(random.uniform(30, 100), 1)
        elif self.type == "motion":
            # Randomly picks True (motion detected) or False (no motion)
            self.current_value = random.choice([True, False])
        elif self.type == "door":
            # Randomly picks OPEN or CLOSED
            self.current_value = random.choice(["OPEN", "CLOSED"])
        elif self.type == "smoke":
            # Randomly picks DETECTED or CLEAR
            self.current_value = random.choice(["DETECTED", "CLEAR"])
        # Return the reading so we can use it immediately
        return self.current_value
    
    # Checks if the current reading is a problem
    def isAbnormal(self):
        if self.type == "temperature":
            # Too cold or too hot = abnormal
            return self.current_value < 35 or self.current_value > 95
        elif self.type == "motion":
            # Any motion = abnormal
            return self.current_value == True
        elif self.type == "door":
            # Door open = abnormal
            return self.current_value == "OPEN"
        elif self.type == "smoke":
            # Smoke detected = abnormal
            return self.current_value == "DETECTED"
        # If sensor type doesn't match anything, it's fine
        return False
    
    # Clears the reading back to nothing
    def reset(self):
        self.current_value = None
    
    # How the sensor looks when you print() it
    def __str__(self):
        status = "No reading" if self.current_value is None else str(self.current_value)
        return f"{self.id} ({self.location}): {status}"
```

```python
import time

def run_simulation(duration_minutes=5, system_mode="AWAY"):
    # Print a nice header
    print("=" * 50)
    print("=== HomeGuard Security System ===")
    print("=" * 50)
    print(f"Mode: {system_mode}\n")
    
    # Create all 4 sensors for the house
    sensors = [
        Sensor("MOTION_001", "Living Room", "motion"),
        Sensor("TEMP_001", "Kitchen", "temperature", threshold=35),
        Sensor("DOOR_001", "Front Door", "door"),
        Sensor("SMOKE_001", "Bedroom", "smoke")
    ]
    
    # Loop once per "minute" of simulation
    for minute in range(duration_minutes):
        # Get the actual current time for display
        current_time = datetime.now().strftime("%H:%M:%S")
        print(f"\nTime: {current_time}")
        
        # Go through each sensor one by one
        for sensor in sensors:
            # Generate a new reading
            reading = sensor.read()
            
            # Display the reading differently based on sensor type
            if sensor.type == "temperature":
                status = "Normal" if 65 <= reading <= 75 else "Abnormal"
                print(f"[READING] {sensor.location} Temperature: {reading}°F ({status})")
            elif sensor.type == "motion":
                status = "DETECTED" if reading else "No activity"
                print(f"[READING] {sensor.location} Motion: {status}")
            elif sensor.type == "door":
                print(f"[READING] {sensor.location}: {reading}")
            elif sensor.type == "smoke":
                print(f"[READING] {sensor.location} Smoke: {reading}")
            
            # If something is wrong, trigger an alert
            if sensor.isAbnormal():
                if sensor.type == "smoke":
                    # Smoke is always critical no matter what mode
                    trigger_alert(create_alert("CRITICAL", "Smoke detected!", sensor.id, current_time))
                elif sensor.type == "motion" and system_mode == "AWAY":
                    # Motion only matters if nobody is home
                    trigger_alert(create_alert("HIGH", "Motion while away!", sensor.id, current_time))
                elif sensor.type == "door" and system_mode in ["AWAY", "SLEEP"]:
                    # Door opening matters when away or sleeping
                    trigger_alert(create_alert("HIGH", "Door opened!", sensor.id, current_time))
        
        # Wait half a second so output doesn't fly by too fast
        time.sleep(0.5)
    
    # Print a nice footer
    print("\n" + "=" * 50)
    print("Simulation complete!")
    print("=" * 50)
    
```

```python
# Start the simulation: 3 minutes, nobody home
run_simulation(duration_minutes=3, system_mode="AWAY")
```
